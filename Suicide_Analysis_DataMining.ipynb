### **Data Cleaning**
<p align = "justify">This code snippet fills missing values in the dataset. For categorical columns (data type 'object'), missing values are replaced with the mode (the most common value), and for numerical columns, with the median. This method helps maintain the integrity of the data distributions.</p>

This step converts selected columns to 'category' data type, which is more memory-efficient and is useful for analysis, especially when encoding categorical variables for machine learning models.

```python
fill_values = {col: df[col].mode()[0] if df[col].dtype == 'object' else df[col].median() for col in df.columns}

df.fillna(fill_values, inplace=True)
```

### **4.1 Model Development - Data Preprocessing and Feature EngineeringVariable**
 **Definitions and Usage:**
In my analysis, We used a combination of demographic and economic variables to predict suicide rates expressed as death rates per 100,000 population. The variables included:

**Categorical Variables:**

**Sex:** This variable represents the gender of the individuals and is crucial for capturing gender-specific trends in suicide rates.

**AgeGroup:**
 Age groups segment the data into coherent demographic units, which helps in understanding at which stages of life suicide rates peak or decline.

**Generation:**
 This categorizes individuals into generational cohorts (e.g., Millennials, Baby Boomers), which provides insights into historical and generational impacts on suicide trends.

**Numerical Variables:**

**Year:**
This helps in analyzing temporal trends and the impact of time-related socio-economic changes on suicide rates.
Population: Used as a base to calculate accurate standardized suicide rates per 100,000 individuals.

**GDPPerCapita:**
 An economic indicator that provides a proxy for the economic status of a region or country, which is often linked to health outcomes including mental health.

<p align = "justify">In the data preprocessing phase for the suicide rates analysis, categorical variables such as Sex, AgeGroup, and Generation are transformed using a OneHotEncoder within a pipeline to handle categorical data efficiently. This encoder converts categorical values into a series of binary columns, ensuring the model can interpret these without any ordinal assumptions. For numerical variables like Year, Population, and GDPPerCapita, no specific transformation is applied within the shown code, but these are directly used as features. The preprocessing steps are centralized through a ColumnTransformer, which applies the encoding to only the categorical columns. This setup simplifies the transformation process, ensuring all data is appropriately formatted for machine learning models. Finally, the transformed data is combined into a single feature matrix X_prepared, ready for model training and analysis.</p>

**We defined the feature matrix X by combining both numerical and categorical columns and set the target variable y as DeathRatePer100K.**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Define columns by types
categorical_columns = ['Sex', 'AgeGroup', 'Generation']
numerical_columns = ['Year', 'Population', 'GDPPerCapita']

# Define transformer for categorical columns
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine transformers into a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_columns)
    ])

# Preprocess the data: Apply transformations
X = df[numerical_columns + categorical_columns]
y = df['DeathRatePer100K']  # Directly use without imputation

# Applying the preprocessing on the feature matrix
X_prepared = preprocessor.fit_transform(X)

X_prepared, y.head()
```

**Splitting the Data:**
The dataset was divided into training and testing sets using train_test_split from scikit-learn, with 20% of the data allocated as the test set. This split ensures that the model can be evaluated on unseen data, providing a more accurate assessment of its performance.

### **Linear Regression Model:**
<p align = "justify">A Linear Regression model was developed as a baseline to predict suicide rates. This model is straightforward and provides a clear understanding of how input features impact the target variable. The model was trained using the training set and then used to predict suicide rates on the test set.

**Performance Metrics:**

**R-squared (R²):** For the Linear Regression model, R² was calculated to be approximately 0.241, indicating that about 24.1% of the variance in the suicide rates is explained by the model.

**Mean Squared Error (MSE):** The MSE for the Linear Regression was approximately 328.61, which quantifies the average squared difference between the observed actual outcomes and the outcomes predicted by the model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_prepared, y, test_size=0.2, random_state=42)

# Create and train the linear regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Predict on the testing set
y_pred = linear_model.predict(X_test)

# Calculate R-squared and Mean Squared Error
r2_lr = r2_score(y_test, y_pred)
mse_lr = mean_squared_error(y_test, y_pred)

r2_lr, mse_lr
```

**Gradient Boosting Regressor:**
To improve prediction accuracy, a Gradient Boosting Regressor model was also employed. This model is more sophisticated and can capture complex nonlinear relationships that the Linear Regression may miss. It was trained using the same training set.

**Gradient Boosting Performance:**

**R-squared (R²):** The R² for the Gradient Boosting model was about 0.297, suggesting it was slightly more effective at explaining the variance in suicide rates compared to the Linear Regression model.

**Mean Squared Error (MSE):** The MSE decreased to about 304.30 with the Gradient Boosting model, indicating a reduction in prediction error compared to the Linear Regression.


```python
from sklearn.ensemble import GradientBoostingRegressor

# Create and train the Gradient Boosting Regressor model
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train, y_train)

# Predict on the testing set using the trained model
y_pred_gb = gb_model.predict(X_test)

# Calculate R-squared and Mean Squared Error for the Gradient Boosting model
r2_gb = r2_score(y_test, y_pred_gb)
mse_gb = mean_squared_error(y_test, y_pred_gb)

r2_gb, mse_gb
```

To visually compare the performance of both models, I plotted their R² values and MSE. The plots clearly showed that while both models performed somewhat similarly, the Gradient Boosting Regressor demonstrated a better fit and lower errors, suggesting its superiority in handling this specific prediction task.

```python
models = ['Linear Regression', 'Gradient Boosting']
r_squared = [r2_lr, r2_gb]
mse = [mse_lr, mse_gb]

# Plotting R^2 values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(models, r_squared, color='blue')
plt.title('R^2 values of Models')
plt.ylabel('R^2 Score')

# Plotting MSE values
plt.subplot(1, 2, 2)
plt.bar(models, mse, color='red')
plt.title('MSE of Models')
plt.ylabel('Mean Squared Error')
plt.show()
```
